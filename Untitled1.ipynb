```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# load dataset
data = pd.read_csv("Sales Dataset.csv")

# display first rows
print(data.head())

# create day column
data['Day'] = range(1, len(data)+1)

# input and output
X = data[['Day']]
y = data['Item_Outlet_Sales']

# create model
model = LinearRegression()

# train model
model.fit(X, y)

# prediction
future_day = [[len(data)+1]]
prediction = model.predict(future_day)

print("Next Day Sales Prediction:", prediction)

# graph
plt.scatter(X, y)
plt.plot(X, model.predict(X))
plt.xlabel("Day")
plt.ylabel("Sales")
plt.title("Sales Forecasting")
plt.show()
```
